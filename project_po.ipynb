```python
#Create the N-1 scenario

#Setting the limits 

v_max = 1.05 #pu
v_min = 0.95 #pu
max_ll = 100 # max load of line
critical_lines = list()

lines = net.line.index # "List of line index"
#print(net.line.in_service)
#pp.runpp(net)
#print(net.line.in_service)
#Checking load of trafo
print("Trafo Before N-1",net.res_trafo.loading_percent)

#Checking load of lines 
#print("Before N-1",net.res_line.loading_percent)
previous = net.res_line.loading_percent

for l in lines:
    net.line.loc[l,"in_service"] = False
    pp.runpp(net)
    
    if net.res_bus.vm_pu.max() > v_max or net.res_bus.vm_pu.min() < v_min or net.res_line.loading_percent.max() > max_ll:
        critical_lines.append(l)
        after = net.res_line.loading_percent
        print(after, "Line in trouble: ", l)

    net.line.loc[l,"in_service"] = True

print("Critical lines after N-1 analysis \n",critical_lines)

ax = pplt.simple_plot(net, show_plot=False)
clc = pplt.create_line_collection(net,critical_lines, color = "r", linewidth=3., use_bus_geodata=True)
pplt.draw_collections([clc],ax=ax)
plt.show()
```

```python
## Function To Analyse the network

def run_analysis(network, v_max, v_min, line_max):
    "Function to run the analysis N-1 of the entire net (network) with the limits v_max, v_min and line_max"

    critical_lines = list()
    critical_buses = list()
    lines = network.line.index # "List of line index"
    buses = network.bus.index # List of buses

    for l in lines:
      net.line.loc[l,"in_service"] = False
      pp.runpp(network)
      
      if network.res_bus.vm_pu.max() > v_max or network.res_bus.vm_pu.min() < v_min or network.res_line.loading_percent.max() > line_max:
          critical_lines.append(l)
          critical_buses.append([network.line.loc[l].from_bus, network.line.loc[l].to_bus])
          print("Line in trouble: ", l)

      net.line.loc[l,"in_service"] = True

    return critical_lines, critical_buses # Returns the list of lines and buses with trouble

lines_in_trouble, buses_in_trouble = run_analysis(net, v_max=1.1,v_min=0.90,line_max=90)

print("Lines in trouble: \n", lines_in_trouble, "\n Buses associated with each line: \n", buses_in_trouble)
pf_res_plotly(net) # To plot results of the Analysis.
```
