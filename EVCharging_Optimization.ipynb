```python
import pandapower as pp
import pandapower.networks as nw
import pandapower.plotting as pplt
import matplotlib.pyplot as plt
net = nw.create_cigre_network_mv()
```

```python
#Create the N-1 scenario

#Setting the limits 

v_max = 1.02 #pu
v_min = 0.9 #pu
max_ll = 100 # max load of line
critical_lines = list()

lines = net.line.index # "List of line index"
#print(net.line.in_service)
#pp.runpp(net)
#print(net.line.in_service)
#Checking load of trafo
print("Before N-1",net.res_trafo.loading_percent)

#Checking load of lines 
print("Before N-1",net.res_line.loading_percent)

for l in lines:
    net.line.loc[l,"in_service"] = False
    pp.runpp(net)
    
    if net.res_bus.vm_pu.max() > v_max or net.res_bus.vm_pu.min() < v_min or net.res_line.loading_percent.max() > max_ll:
        critical_lines.append(l)
    net.line.loc[l,"in_service"] = True
    print(net.res_trafo.loading_percent, "Line in trouble: ", l)

print("critical lines after N-1 analysis",critical_lines)

ax = pplt.simple_plot(net, show_plot=False)
clc = pplt.create_line_collection(net,critical_lines, color = "r", linewidth=3., use_bus_geodata=True)
pplt.draw_collections([clc],ax=ax)
plt.show()
```
